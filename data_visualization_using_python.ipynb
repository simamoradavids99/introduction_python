```python
# Group data by 'Pclass' and calculate the survival rate for each class
survival_rate_by_class = titanic_df.groupby('Pclass')['Survived'].mean()

# Group data by 'Pclass' and calculate the average age for each class
average_age_by_class = titanic_df.groupby('Pclass')['Age'].mean()
```

```python
survival_rate_by_class = titanic_df.groupby('Pclass')['Survived'].mean()
```

```python
average_age_by_class = titanic_df.groupby('Pclass')['Age'].mean()
```

```python
survival_rate_by_class
```

```python
# Plotting the survival rate by class
plt.figure(figsize=(8, 6))
sns.barplot(x=survival_rate_by_class.index, y=survival_rate_by_class.values)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
plt.figure(figsize = (8,6))
sns.barplot(x = survival_rate_by_class.index, y=survival_rate_by_class.values)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
# Plotting the average age by class
plt.figure(figsize=(8, 6))
sns.barplot(x=average_age_by_class.index, y=average_age_by_class.values)
plt.title('Average Age by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Average Age')
plt.show()
```
